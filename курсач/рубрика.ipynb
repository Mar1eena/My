```python
#рандомные данные для проверки работы нейронной сети
price=np.random.randint([50,], 731,size=(1224,1))
totsp=np.random.randint([44,], 193,size=(1224,1))
livesp=np.random.randint([28,], 103,size=(1224,1))
kitsp=np.random.randint([5,], 26,size=(1224,1))
dist=np.random.randint([3,], 18,size=(1224,1))
metrdist=np.random.randint([1,], 21,size=(1224,1))
walk=np.random.randint([0,], 2,size=(1224,1))
brick=np.random.randint([0,], 2,size=(1224,1))
floor=np.random.randint([0,], 2,size=(1224,1))
random_data=np.hstack((price,totsp,livesp,kitsp,dist,metrdist,walk,brick,floor))
random_data = pd.DataFrame(random_data,columns=[['price','totsp','livesp','kitsp','dist','metrdist','walk','brick','floor']])
```
